### Standard Imports

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.math import confusion_matrix

from tensorflow.keras.utils import plot_model
import tensorflow as tf
```

### 1. Data Preparation and Encoding

#### 1.1 Import Modified Data

```python
# Import the Student's data from local storage
# ! Dont' run the cell in colab
data_df = pd.read_csv('./data/xAPI-Edu-Data_modified.csv')
```

```python
# Import the Student's data from github storage
# Run only in colab
url = 'https://raw.githubusercontent.com/AamirKhaan/Student-Academic-Performance/main/data/xAPI-Edu-Data_modified.csv'
data_df = pd.read_csv(url)
```

```python
# Selected Features
CATEGORICAL_FEATURES = ['stage_id','relation', 'parent_answering_survey', 
                            'parent_school_satisfaction','student_absence_days']
NUMERICAL_FEATURES = ['raised_hands', 'visited_resources', 'announcements_view', 'discussion']
TARGET = ['class']
```

```python
# Optimize Data Type for efficent memory utilization
for feature in CATEGORICAL_FEATURES:
    data_df[feature] = data_df[feature].astype('category')
    
for feature in NUMERICAL_FEATURES:
    data_df[feature] = data_df[feature].astype('int8')

data_df[TARGET] = data_df[TARGET].astype('category')
```

#### 1.2 Encode Data

```python
# Input Features Encoding
categorical_encoder = OrdinalEncoder(dtype='int8')
categorical_ds  = categorical_encoder.fit_transform(data_df[CATEGORICAL_FEATURES])
categorical_encoded = pd.DataFrame(categorical_ds, columns=CATEGORICAL_FEATURES, dtype='int8')
# Target Encoding
target_encoder = LabelEncoder()
target_ds = target_encoder.fit_transform(data_df[TARGET])
target_encoded = pd.DataFrame(target_ds, columns=TARGET, dtype='int8')
# Combine the Data
encoded_df = pd.concat([categorical_encoded,data_df[NUMERICAL_FEATURES]], axis=1)
```

### 2.  Preliminary DNN Model

```python
X = encoded_df
y = target_encoded
```

```python
# Test train split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Standard Scaler Transform
transformer = StandardScaler()
transformer.fit(X_train)

X_train = transformer.transform(X_train)
X_test = transformer.transform(X_test)
```

```python
# Random Seed 
tf.random.set_seed(42)
```

```python
# DNN Model
dnn_model= Sequential()
dnn_model.add(Dense(64, input_shape = (X_train.shape[1],), activation ='relu', name = 'Deep_Layer_1'))
dnn_model.add(Dense(32, activation ='relu', name = 'Deep_Layer_2'))
dnn_model.add(Dense(16, activation ='relu', name = 'Deep_Layer_3'))
dnn_model.add(Dense(8, activation ='relu', name = 'Deep_Layer_4'))
dnn_model.add(Dense(3, activation='softmax', name = 'Output_Layer'))
```

```python
# Loss and Optimizer
loss = tf.keras.losses.SparseCategoricalCrossentropy()
optimizer = tf.keras.optimizers.RMSprop(0.0001)
```

```python
# Compile the Model
dnn_model.compile(optimizer=optimizer,
              loss=loss,
              metrics=['accuracy'])
```

#### 2.1 Model Parameters

```python
dnn_model.summary()
```

#### 2.2 DNN Structure

```python
plot_model(dnn_model, to_file='images/DNN_model.png', show_shapes=True)
```

```python
result = dnn_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=1000, verbose=0);
```

#### 2.3 Model Scores

```python
# Test Accuracy 
print(f"Preliminary DNN Test Accuracy: {result.history['val_accuracy'][-1]*100}")
# Train Accuracy
print(f"Preliminary DNN Train Accuracy: {result.history['accuracy'][-1] *100}")
```

<div style="background-color:Gainsboro; padding:20px; text-align:justify; ">
<p>The accuracy of the model on Training Data (96.875%) and the accuracy on Test Data (75.316%).</p>
</div>

#### 2.4 Preliminary DNN Learning Curve

```python
# Plot loss per epoch
plt.plot(result.history['loss'], label='Training Loss')
plt.plot(result.history['val_loss'], label='Test Loss')
plt.title('Learning Curves')
plt.xlabel('Epochs')
plt.ylabel('Cross Entropy Loss')
plt.legend();
```

#### 2.5 Preliminary DNN Accuracy Curve

```python
# Plot accuracy per epoch
plt.plot(result.history['accuracy'], label='Training Accuracy')
plt.plot(result.history['val_accuracy'], label='Test Accuracy')
plt.title('Accuracy Curves')
plt.xlabel('Epochs')
plt.ylabel('DNN Accuracy')
plt.legend();
```

<div style="background-color:Gainsboro; padding:20px; text-align:justify;">
<p>The graph of DNN Accuracy vs Epochs show the variation in the accuracy with respect to Epochs in Training and Test data. The highest accuracy in test data is when the number of epochs are around 500.</p>
</div>

#### 2.6 Confusion Matrix

```python
y_hat = [np.argmax(x) for x in dnn_model.predict(X_test)]

dnn_cm_df = pd.DataFrame(confusion_matrix(y_test, y_hat),
                     index = ['H','L','M'], 
                     columns = ['H','L','M'])

fig = plt.figure(figsize=(6,6))
sns.heatmap(dnn_cm_df, annot=True, cmap='mako',cbar=False,linewidth=1)
plt.title('Preliminary DNN Confusion Matrix')
plt.ylabel('Actual Values')
plt.xlabel('Predicted Values')
plt.show()
```

#### 2.7 Classification Report

```python
from sklearn.metrics import classification_report
print('Preliminary DNN Classification Report\n')
print(classification_report(y_test,y_hat, target_names=['H','L','M']))
```

### 3. DNN with optimized Number of Epochs

<div style="background-color:Gainsboro; padding:20px; text-align:justify;">
The best model can be found out by seraching for the maximum test accuracy in the history.
</div>

```python
max_test_accuracy = max(result.history['val_accuracy'])*100
print(f'Maximum Test Accuracy : {max_test_accuracy}')
```

```python
optimal_epoch = np.where(np.array(result.history['val_accuracy']) == max(result.history['val_accuracy']))
optimal_epoch
```

<div style="background-color:Gainsboro; padding:20px; text-align:justify;">
Any suitable epoch value can be selected for the ANN training from the list of optimal_epoch.
</DIV>

```python
# Random Seed 
tf.random.set_seed(42)

# DNN Model
opt_dnn_model= Sequential()
opt_dnn_model.add(Dense(64, input_shape = (X_train.shape[1],), activation ='relu', name = 'Deep_Layer_1'))
opt_dnn_model.add(Dense(32, activation ='relu', name = 'Deep_Layer_2'))
opt_dnn_model.add(Dense(16, activation ='relu', name = 'Deep_Layer_3'))
opt_dnn_model.add(Dense(8, activation ='relu', name = 'Deep_Layer_4'))
opt_dnn_model.add(Dense(3, activation='softmax', name = 'Output_Layer'))

# Loss and Optimizer
loss = tf.keras.losses.SparseCategoricalCrossentropy()
optimizer = tf.keras.optimizers.RMSprop(0.0001)

# Compile the Model
opt_dnn_model.compile(optimizer=optimizer,
              loss=loss,
              metrics=['accuracy'])

opt_result = opt_dnn_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=500, verbose=0);
```

#### 3.1 Optimized DNN Scores

```python
# Test Accuracy 
print(f"Optimized DNN Test Accuracy: {opt_result.history['val_accuracy'][-1]*100}")
# Train Accuracy
print(f"Optimized DNN Train Accuracy: {opt_result.history['accuracy'][-1] *100}")
```

<div style="background-color:Gainsboro; padding:20px; text-align:justify; ">
<p>The accuracy of the optimized model on Training Data (89.99%) is close to the accuracy on Test Data (76.58%). The model shows acceptable performance. </p>
</div>

#### 3.2 Optimized DNN Learning Curve

```python
# Plot loss per epoch
plt.plot(opt_result.history['loss'], label='Training Loss')
plt.plot(opt_result.history['val_loss'], label='Test Loss')
plt.title('Learning Curves')
plt.xlabel('Epochs')
plt.ylabel('Cross Entropy Loss')
plt.legend();
```

#### 3.3 Optimized DNN Accuracy Curve

```python
# Plot accuracy per epoch
plt.plot(opt_result.history['accuracy'], label='Training Accuracy')
plt.plot(opt_result.history['val_accuracy'], label='Test Accuracy')
plt.title('Accuracy Curves')
plt.xlabel('Epochs')
plt.ylabel('DNN Accuracy')
plt.legend();
```

#### 3.4 Optimized DNN Confusion Matrix

```python
opt_y_test = [np.argmax(x) for x in opt_dnn_model.predict(X_test)]
ocm_df = pd.DataFrame(confusion_matrix(y_test, opt_y_test),
                     index = ['H','L','M'], 
                     columns = ['H','L','M'])

fig = plt.figure(figsize=(6,6))
sns.heatmap(ocm_df, annot=True, cmap='mako',cbar=False,linewidth=1)
plt.title('Optimized DNN Confusion Matrix')
plt.ylabel('Actual Values')
plt.xlabel('Predicted Values')
plt.show()
```

#### 3.5 Optimized DNN Classification report

```python
print('Optimized DNN Classification Report\n')
print(classification_report(y_test,opt_y_test, target_names=['H','L','M']))
```
